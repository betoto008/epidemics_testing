```python
from scipy.optimize import curve_fit
from scipy.interpolate import interp1d
import sys
sys.path.append('../Codes/')
from models import *
from functions import *
import networkx as nx
import matplotlib.animation as animation
def node_degrees(Amat):
        return Amat.sum(axis=0).reshape(N,1)
```

## Random Shell Network

```python
N =200
emp = 160
pat = N-emp
constructor = [(emp, N, (1)/(2)), (pat, pat, 1/10)]
G = nx.random_shell_graph(constructor)
A = networkx.adj_matrix(G)
#print(A)
fig, ax = plt.subplots(1,2,figsize=(16,8))
nx.draw_shell(G, nlist = [range(emp), range(emp, pat+emp)], node_color=['g']*emp+['r']*pat, style ='dashed',  with_labels=True, width = 0.5, ax=ax[0])
ax[1].spy(A, origin = 'lower', ms = 0.5)
ax[1].hlines(emp+0.5, -0.5, N, color = 'k')
ax[1].vlines(emp+0.5, -0.5, N, color = 'k')
```

```python
print(type(node_degrees(networkx.adj_matrix(G))))
a = np.asarray(node_degrees(networkx.adj_matrix(G))).astype(float)
plot_degree_distn(G, max_degree=10)
```
